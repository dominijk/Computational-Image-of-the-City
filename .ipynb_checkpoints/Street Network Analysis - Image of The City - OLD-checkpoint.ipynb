# Functions

```python
def straightness_centrality(G, w, normalized=True, distance=True):

    if distance is True:
        distance=w
    
    
    path_length = functools.partial(nx.single_source_dijkstra_path_length, weight=distance)

    nodes = G.nodes()
    straightness_centrality = {}
    # TODO one request to get all the coordinates only
    # edges_coordinates = init_coordinates()

    # Initialize dictionary containing all the node id and coordinates
    # coord_nodes = get_nodes_coords(Node, Session)
    coord_nodes = nodes_dict(G)

    for n in nodes:
        straightness = 0
        sp = path_length(G,n)

        if len(sp) > 0 and len(G) > 1:
            # start computing the sum of euclidean distances
            source = {'x':coord_nodes.get(n)[0],'y':coord_nodes.get(n)[1]}
            # if the source is not a virtual node
            if source is not None:
                for target in sp:
                    if n != target and target in coord_nodes:
                        network_dist = sp[target]

                        euclidean_dist = euclidean_distance(*coord_nodes[n]+coord_nodes[target])
                        straightness = straightness + (euclidean_dist/network_dist)

                straightness_centrality[n]= straightness
                if normalized:
                        straightness_centrality[n]= straightness * (1.0/(len(G)-1.0) )

        else:               
            straightness_centrality[n]=0.0

    return straightness_centrality


def euclidean_distance(xs, ys, xt, yt):
    """ xs stands for x source and xt for x target """
    return sqrt((xs - xt)**2 + (ys - yt)**2)

def nodes_dict(G):
    nodes_list = G.nodes(1)
    look = {}

    for i in range(0, len(nodes_list)):
        cod = nodes_list[i][0]
        x = nodes_list[i][1]['x']
        y = nodes_list[i][1]['y']
        look[cod] = (x,y)
    
    return look

def edge_straightness(G, Cs):
    edge_st = {}
    t = G.edges()
    for i in t:
         edge_st[i] = (nodes_df.get_value(i[0], Cs)+nodes_df.get_value(i[1], Cs))/2
    
    return edge_st

import math
def dot(vA, vB):
    return vA[0]*vB[0]+vA[1]*vB[1]
def ang(lineA, lineB):
    # Get nicer vector form
    vA = [(lineA[0][0]-lineA[1][0]), (lineA[0][1]-lineA[1][1])]
    vB = [(lineB[0][0]-lineB[1][0]), (lineB[0][1]-lineB[1][1])]
    # Get dot prod
    dot_prod = dot(vA, vB)
    # Get magnitudes
    magA = dot(vA, vA)**0.5
    magB = dot(vB, vB)**0.5
    # Get cosine value
    cos_ = dot_prod/magA/magB
    # Get angle in radians and then convert to degrees
    angle = math.acos(dot_prod/magB/magA)
    # Basically doing angle <- angle mod 360
    ang_deg = math.degrees(angle)%360
    return ang_deg
     #if ang_deg-180<=0:
     #    return 180 - ang_deg
    # else:
     #    return ang_deg
        

        



def reach(G, w, radius, normalized=True, distance=True):

    if distance is True:
        distance=w
    
    path_length = functools.partial(nx.single_source_dijkstra_path_length, weight=distance)

    nodes = G.nodes()
    reach_centrality = {}
    coord_nodes = nodes_dict(G)

    for n in nodes:
        reach = 0
        sp = path_length(G,n)
        sp_radium = dict((k, v) for k, v in sp.items() if v <= radius)
        
        if len(sp_radium) > 0 and len(G) > 1:
           
            source = {'x':coord_nodes.get(n)[0],'y':coord_nodes.get(n)[1]}
            # if the source is not a virtual node
            if source is not None:
                for target in sp_radium:
                    if n != target and target in coord_nodes:
                        weight_target = G.node[target]['weight']
                        reach = reach + weight_target
                        

                reach_centrality[n] = reach

        else:               
            reach_centrality[n]=0.0

    return reach_centrality

def ang_rad(lineA, lineB):
    # Get nicer vector form
    vA = [(lineA[0][0]-lineA[1][0]), (lineA[0][1]-lineA[1][1])]
    vB = [(lineB[0][0]-lineB[1][0]), (lineB[0][1]-lineB[1][1])]
    # Get dot prod
    dot_prod = dot(vA, vB)
    # Get magnitudes
    magA = dot(vA, vA)**0.5
    magB = dot(vB, vB)**0.5
    # Get cosine value
    cos_ = dot_prod/magA/magB
    # Get angle in radians and then convert to degrees
    angle = math.acos(dot_prod/magB/magA)

    return angle


def local_CB(G, w, radius, distance=True):

    if distance is True:
        distance=w
        
    path_length = functools.partial(nx.single_source_dijkstra_path_length, weight=distance)

    nodes = G.nodes()
    cb = {}
    coord_nodes = nodes_dict(G)

    for obj in nodes:
        sp = path_length(G,obj)
        sp_radium = dict((k, v) for k, v in sp.items() if v <= radius)

        to_keep = list(sp_radium.keys())
        G_small = G.subgraph(to_keep)
        
        for i in G_small.edges():
            G_small.edge[i[0]][i[1]]['distance']=G_small.edge[i[0]][i[1]][0]['distance']


        be = nx.betweenness_centrality(G_small, k=None, weight = 'distance', normalized=False)
        cb[obj] = be[obj]
        
    return cb
```

# Dual Analysis

```python
# centroids for dual analysis

centroids_df = result.copy()
centroids_df['centroid'] = gpd.GeoSeries(centroids_df['geometry'].centroid)
```

```python
centroids_df['intersecting'] = "NaN"
centroids_df['intersecting'] = centroids_df['intersecting'].astype(object)
```

```python
#find_intersecting

processed = []

for e in centroids_df.itertuples():
    intersections = []
        
    from_node = e[6] #check before running the right column of "u" and "v"
    to_node = e[7]
    
    tmp = centroids_df.loc[(centroids_df['u'] == from_node) |
                        (centroids_df['u'] == to_node) |
                        (centroids_df['v'] == to_node) |
                        (centroids_df['v'] == from_node)]
    
    for l in tmp.itertuples():
        if ((e[0]==l[0]) | (((e[0], l[0]) in processed) == True)
            | (((l[0], e[0]) in processed) == True)) :
            continue
        else:
            intersections.append(l[1])  #appending OBJECTID
            processed.append((l[0],e[0]))
    
    centroids_df.set_value(e[0],'intersecting', intersections)
```

```python
#creating nodes representing street segments
nodes_centroids_data = centroids_df[['OBJECTID', 'intersecting', 'distance']]
crs = {'init': 'epsg:27700'}
geometry = centroids_df['centroid']
```

```python
nodes_centroids = gpd.GeoDataFrame(nodes_centroids_data, crs=crs, geometry=geometry)
nodes_centroids['x'] = [x.coords.xy[0][0] for x in centroids_df['centroid']]
nodes_centroids['y'] = [y.coords.xy[1][0] for y in centroids_df['centroid']]
```

```python
#storing final "list" of nodes
nodes_cn = nodes_centroids.drop('intersecting', axis=1)
nodes_cn.gdf_name = 'Dual_list'
```

```python
junctions = pd.DataFrame(columns=['u','v', 'key', 'geometry', 'distance'])
```

```python
for row in nodes_centroids.itertuples():
        
    obj_id = row[1] #object_id of the relative segment
    length = row[3]
        
    for i in list(row[2]): #intersecting segments
        # i is the ObjectID

        index = centroids_df.index[centroids_df.OBJECTID==i].tolist()[0]
        length_v =  centroids_df['distance'][centroids_df.OBJECTID==i][index]
        distance = (length+length_v)/2
        
         # adding a row with u-v, key fixed as 0, Linestring geometry from the first centroid to centroid intersecting segment 
        
        junctions.loc[-1] = [obj_id, i, 0,  
                            LineString([row[4], nodes_centroids.loc[index]['geometry']]),
                            distance] 
          
        junctions.index = junctions.index + 1
```

```python
#creating geodf

junctions = junctions.sort_index(axis=0)
crs = {'init': 'epsg=2806'}
geometry = junctions['geometry']
junctions = gpd.GeoDataFrame(junctions[['u','v', 'key', 'distance']], crs=crs, geometry=geometry)
```

```python
#computing angles

for row in junctions.itertuples():

    #retrieveing original lines from/to
    
    from_node = nodes_geo.loc[result['u'].loc[result.OBJECTID==row[1]]].index.tolist()[0]
    to_node = nodes_geo.loc[result['v'].loc[result.OBJECTID==row[1]]].index.tolist()[0]
                                            
    from_node2 = nodes_geo.loc[result['u'].loc[result.OBJECTID==row[2]]].index.tolist()[0]           
    to_node2 = nodes_geo.loc[result['v'].loc[result.OBJECTID==row[2]]].index.tolist()[0]
    
    if ((from_node == from_node2) & (to_node == to_node2) |
        (from_node == to_node2) & (to_node == from_node2)):
        deflection = 0
    
    else:
         
        try:  
        
            x_f = float("{0:.10f}".format(nodes_geo.loc[from_node]['x']))
            y_f = float("{0:.10f}".format(nodes_geo.loc[from_node]['y']))
            x_t = float("{0:.10f}".format(nodes_geo.loc[to_node]['x']))
            y_t = float("{0:.10f}".format(nodes_geo.loc[to_node]['y']))
            
            x_f2 = float("{0:.10f}".format(nodes_geo.loc[from_node2]['x']))
            y_f2 = float("{0:.10f}".format(nodes_geo.loc[from_node2]['y']))    
            x_t2 = float("{0:.10f}".format(nodes_geo.loc[to_node2]['x']))
            y_t2 = float("{0:.10f}".format(nodes_geo.loc[to_node2]['y']))          
                             
            if (to_node == to_node2):
                lineA = ((x_f, y_f),(x_t,y_t))
                lineB = ((x_t2, y_t2),(x_f2, y_f2))
    
            elif (to_node == from_node2):
                lineA = ((x_f, y_f),(x_t,y_t))
                lineB = ((x_f2, y_f2),(x_t2, y_t2))

            elif (from_node == from_node2):
                lineA = ((x_t, y_t),(x_f,y_f))
                lineB = ((x_f2, y_f2),(x_t2, y_t2))

            else: #(from_node == to_node2)
                lineA = ((x_t, y_t),(x_f,y_f))
                lineB = ((x_t2, y_t2),(x_f2, y_f2))
        
            deflection = ang(lineA, lineB)
            deflection_rad = ang_rad(lineA, lineB)
        except:
            deflection = 0
            deflection_rad = 0
    
    junctions.set_value(row[0],'angle', deflection)
    junctions.set_value(row[0],'angle_rad', deflection_rad)
```

```python
def get_node_colors_by_stat(G, data, start=0, end=1):
    df = pd.DataFrame(data=pd.Series(data).sort_values(), columns=['value'])
    df['colors'] = get_color_list(len(df), start=start, end=end)
    df = df.reindex(G.nodes())
    return df['colors'].tolist()
```
