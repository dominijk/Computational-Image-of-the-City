```python
import pandas as pd
import numpy as np
import pandana
import geopandas as gpd
import geopandas.io
import geopandas_osm.osm as osm
from shapely import geometry
import osmnx as ox
```

```python
%matplotlib inline
```

```python
G = ox.graph_from_place('Torino', network_type = 'all', simplify=True)
G = ox.project_graph(G)
```

```python
ox.plot_graph(G, fig_height=19)
```

```python
ox.save_graph_shapefile(G, filename='Torino_street_network')
```

```python
gdf_buildings = ox.buildings_from_place(place='Torino')
gdf_buildings = ox.project_gdf(gdf_buildings)
```

```python
gdf_save = gdf_buildings.drop(labels='nodes', axis=1)
gdf_save.to_file('data/Turin_buildings')
```
